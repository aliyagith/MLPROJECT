```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv")
df
```

```python
X =df.iloc[:,:-1]
print(X)
```

```python
y =df.iloc[:,1].values
print(y)
```

Model Training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state= 0)# random =0 esliye likhte hai taaki agar data baar baar run karaye to output different na aaye
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Prediciting The Result

```python
y_pred = regressor.predict(X_test)
```

Making a single prediction(for example the salary of an employees with 12 years of expression)

```python
print(df,regressor.predict([[12]]))
```

Visulaizing the training result

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('salary vs Expression (Training Set)')
plt.xlabel('Year of Expression')
plt.ylabel('salary')
plt.show()
```
