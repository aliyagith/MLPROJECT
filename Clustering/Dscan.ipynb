DBSCAN-density based spatial clustering of application with noise

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler  
```

```python
df = pd.read_csv('mallcustomers.csv')
df.head()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
X = StandardScaler().fit_transform(X)
```

- eps - Neighbourhood radius :how close two point must be to be in the same cluster 
- min_samples- minimum points in a cluster -the minimum  number of point needed to form a cluster


```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
labels =dbscan.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='viridis',s =50)
plt.title('DBSCAN clustering of cluster')
plt.xlabel('Annual Income')
plt.ylabel('spending Score')
plt.show()
```

```python
#finding outliers 
outliers = X[labels == -1]
print(f'total outliers detected:{len(outliers)}')
```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='viridis',s =50)
plt.scatter(outliers[:,0],outliers[:,1],color='red',s=100,label='outliers',marker ='x')
plt.title('DBSCAN clustering of cluster')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
